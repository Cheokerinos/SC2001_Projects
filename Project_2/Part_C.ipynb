```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
Matrix_VaryV = pd.read_csv('Matrix_Fixed_E_Vary_V.csv')

Matrix_VaryV_Comparisons = Matrix_VaryV.drop(columns = ['CPU Time (ms)'])

List_VaryV = pd.read_csv('List_Fixed_E_Vary_V.csv')

List_VaryV_Comparisons = List_VaryV.drop(columns=['CPU Time (ms)'])
List_VaryV_Comparisons.head()
```

```python
plt.figure(figsize=(10,6))

sb.lineplot(x='Number_Of_Vertices', y='Comparisons', data=Matrix_VaryV_Comparisons, marker='o')

plt.title('Matrix: Number of Vertices vs Comparisons (Fixed Edges of 1000)')
plt.xlabel('Number of Vertices')
plt.ylabel('Number of Comparisons')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sb.lineplot(x='Number_Of_Vertices', y='Comparisons', data=List_VaryV_Comparisons, marker='X')

plt.title('List: Number of Vertices vs Comparisons (Fixed Edges of 10000)')
plt.xlabel('Number of Vertices')
plt.ylabel('Number of Comparisons')
plt.grid(True)
plt.show()
```

```python
Matrix_VaryV_Time = Matrix_VaryV.drop(columns = ['Comparisons'])

List_VaryV_Time = List_VaryV.drop(columns=['Comparisons'])

plt.figure(figsize=(10,6))

sb.lineplot(x='Number_Of_Vertices', y='CPU Time (ms)', data=List_VaryV_Time, marker='X')
sb.lineplot(x='Number_Of_Vertices', y='CPU Time (ms)', data=Matrix_VaryV_Time, marker='o')

plt.title('Number of Vertices vs Comparisons (Fixed Edges of 1000)')
plt.xlabel('Number of Vertices')
plt.ylabel('CPU Time (ms)')
plt.grid(True)
plt.show()
```

```python
Matrix_VaryE = pd.read_csv('Matrix_Fixed_V_Vary_E.csv')

Matrix_VaryE_Time = Matrix_VaryE.drop(columns = ['Comparisons'])

List_VaryE = pd.read_csv('List_Fixed_V_Vary_E.csv')

List_VaryE_Time = List_VaryE.drop(columns=['Comparisons'])

List_VaryE_Time.head()
```

```python
plt.figure(figsize=(10,6))

sb.lineplot(x='Number_Of_Edges', y='CPU Time (ms)', data=Matrix_VaryE_Time, marker='o')
sb.lineplot(x='Number_Of_Edges', y='CPU Time (ms)', data=List_VaryE_Time, marker='X')

plt.title('Number of Edges vs CPU Time (Fixed Vertices of 10000)')
plt.xlabel('Number of Edges')
plt.ylabel('CPU Time (ms)')
plt.grid(True)
plt.show()
```
