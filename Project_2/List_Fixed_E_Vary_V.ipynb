Graphing for Adjacency Matrix, with fixed edges (1000) and varying vertices (From 100 to 5000, increment by 100)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the CSV data
data = pd.read_csv("List_Fixed_E_Vary_V.csv")

# Since edges are fixed, create a new column for edges
fixed_edges = 1000  # Use the fixed edge value you set in the C code
data['Number_Of_Edges'] = fixed_edges
```

```python
import matplotlib.pyplot as plt

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(data['Number_Of_Vertices'], data['CPU Time (ms)'], marker='o', color='b')

# Add titles and labels
plt.title('Dijkstra Algorithm: CPU Time vs Number of Vertices (Fixed Edges)')
plt.xlabel('Number of Vertices')
plt.ylabel('CPU Time (ms)')
plt.grid(True)

# Show the plot
plt.show()
```
