Graphing for Adjacency Matrix, with fixed vertices (10000) and varying edges (From 1000 to 100000, increment by 1000)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file generated by your C code
data = pd.read_csv('Matrix_Fixed_V_Vary_E.csv')

# Display the first few rows to verify the data
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Plot the data for Fixed Vertices, Varying Edges
plt.figure(figsize=(10, 6))

# Plot Number of Edges vs CPU Time
plt.plot(data['Number_Of_Edges'], data['CPU Time (ms)'], marker='o', linestyle='-', color='r')

# Add titles and labels
plt.title('Dijkstra Algorithm: CPU Time vs Number of Edges (Fixed Vertices)')
plt.xlabel('Number of Edges')
plt.ylabel('CPU Time (ms)')
plt.grid(True)

# Show the plot
plt.show()
```
