Graphing for Adjacency Matrix, with fixed vertices and varying edges

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file generated by your C code
data = pd.read_csv('Matrix_Fixed_V_Vary_E.csv')

# Display the first few rows to verify the data
print(data.head())
```

Graph shows that the number of comparisons increases as the number of edges increases. However, as you can see, the difference here is very small (5000 per increment on the y-axis), so it's not really too significant, but it still shows that there is an increase. 

There is, however, a jump in comparison of 5377 from 6000 to 7000 edges. From 1000 to 6000 edges, the comparisons are of the same value. 

1e^7 means 10 million

```python
plt.figure(figsize=(10,6))

# Use seaborn for a clean plot
sns.lineplot(x='Number_Of_Edges', y='Comparisons', data=data, marker='o')

plt.title('Number of Edges vs Comparisons in Dijkstra\'s Algorithm')
plt.xlabel('Number of Edges')
plt.ylabel('Number of Comparisons')
plt.grid(True)
plt.show()
```

Here it's just to show the average comparisons from all of the edges. As you can see, the average is very close to the number of comparisons (y-axis) in the 1st graph. So there is no significant jumps in comparison as number of edges increases, basically it does not really make a difference from 7000 edges onwards.

```python
# Calculate the overall average of 'Comparisons'
overall_avg_comparisons = data['Comparisons'].mean()

# Display the average
print(f'Overall Average Comparisons: {overall_avg_comparisons}')

# Set the figure size
plt.figure(figsize=(10,6))

# Plot a straight horizontal line for the overall average
plt.axhline(y=overall_avg_comparisons, color='r', linestyle='--', label=f'Average Comparisons = {overall_avg_comparisons:.2f}')

# Add labels and title
plt.title('Average Comparisons vs Number of Edges')
plt.xlabel('Number of Edges')
plt.ylabel('Average Comparisons')
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
